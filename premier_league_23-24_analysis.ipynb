# 1.Goals Scored by each team in the premier League Season 2023-24
#### This visualization displays the total goals scored by each team during the 2023–24 Premier League season, highlighting differences in attacking output across the league. It allows for quick comparison between the most prolific and least effective attacking sides, offering insights into team strengths, playing styles, and overall offensive performance throughout the season.

```python
teams= df.groupby("Team")
teams_gls=teams[["Gls"]].sum().sort_values(by="Gls",ascending=False)
teams_gls
```

```python
fig,ax=plt.subplots(figsize=(12,12))
sns.barplot(x="Gls",y="Team",data=teams_gls,palette="viridis",hue="Team");
ax.legend(["Goals scored by each team"],loc="lower right")
ax.set_xlabel("Goals scored",color="purple",size=16,font="Bahnschrift")
ax.set_ylabel("Teams",color="purple",size=16,font="Bahnschrift")
ax.set_title("Premier League 23-24 Goals by Teams",font="Bahnschrift",color="purple",size=16)

plt.savefig("Team-Goals")
```

# 2.Top ten Goal scorers of the premier League season 2023-24
#### This visualization presents the top goal scorers of the 2023-24 premier league season, highlighting the players who have consistently delivered exceptional performances in front of goal. The chart compares the number of goals scored by each player.

```python
scorer=df[["Player","Gls","xG"]]
top_scorers=scorer.nlargest(10,"Gls")
top_scorers
```

```python
fig,ax=plt.subplots(figsize=(12,12))
sns.barplot(x="Player",y="Gls",data=top_scorers,hue="Player",palette="Purples",edgecolor="Black")
ax.tick_params(axis="x",rotation=45)
ax.legend(["Goals scored by individuals"])
ax.set_title("Premier League 23-24 top Goal scorers",font="Bahnschrift",size=16,color="purple")
ax.set_xlabel("Players",size=12,color="purple",font="Bahnschrift")
ax.set_ylabel("Goals scored",size=12,color="purple",font="Bahnschrift")

plt.savefig("Top-Ten-Goal-scorers")
```

# 3.Top ten assist providers of the Premier League season 2023-24
#### This visualization showcases the top assist providers of the current football season, focusing on the players who have created the most goal-scoring opportunities for their teammates. The chart compares the total number of assists.

```python
assists=df[["Player","Ast"]]
top_assists=assists.nlargest(10,"Ast")
top_assists
```

# 4.Players with the Most Expected goals
#### Expected Goals (xG) is a key advanced football metric that measures the quality of a scoring chance rather than just counting goals. It assigns a value between 0 and 1 to each shot, representing the probability of it resulting in a goal based on historical data of similar situations. The below data exploration and visualization represents the Players with the most expected goals. The line plot refers to the actual goals scored by that player in this season. Which makes us to judge players whether they are over or under performing.

```python
ex_gls=df[["Player","xG","Gls"]].nlargest(10,"xG")
ex_gls
```

```python
fig,ax= plt.subplots(figsize=(12,12))
bar=sns.barplot(x="Player", y="xG",data=ex_gls,palette="flare",hue="Player")
line=sns.lineplot(x="Player",y="Gls", data=ex_gls, color="blue", markers=True, marker="o")
ax.tick_params(axis="x",rotation= 45)
ax.set_xlabel("Expected goals",font="Bahnschrift",size=12,color="purple")
ax.set_ylabel("Goals scored by the player", font="Bahnschrift", size= 12,color="purple")
ax.legend(["Expected Goals"])
fig.suptitle("premier league 23-24", font="Bahnschrift",size=16,color="purple")
ax.set_title("Players with the most expected Goals vs actual goals scored", font="Bahnschrift",size=16,color="purple")

plt.savefig("Top-Expected-Goals")
```

# 5.Time distribution of the players by different position 
#### This violin plot illustrates the distribution of time spent in the field by players across different positions. The visualization reveals patterns in workload and game involvement, showing how playing time varies within each role. Wider sections of the violin represent more frequent time values, while narrower sections indicate less common durations. By comparing these distributions, coaches, analysts, and fans can identify which positions typically have longer or shorter on-field presence, and assess variations between players in similar roles.


```python
df["pos"]=df["Pos"].replace({"MF,DF":"MF","MF,FW":"MF","DF,MF":"MF","FW,MF":"FW","DF,FW":"DF","FW,DF":"FW"},inplace=True)
```

```python
pos=df.groupby("Pos")
pos["Pos"].value_counts()
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sns.violinplot(x="Pos",y="Min",data=df,palette="pastel",hue="Pos")
ax.set_xlabel("Postion",size=12,font="Bahnschrift",c="#3D195B")
ax.set_ylabel("Number of minutes played",size=12,font="Bahnschrift",c="#3D195B")
ax.legend(["Distribution of Minutes","outliers","Interquartile range","Mean"])
ax.set_title("Distribution of Minutes played by players of each position",font="Bahnschrift",size=16,c="#3D195B")
fig.suptitle("Premier League23-24",font="Bahnschrift",size=16,c="#3D195B")

plt.savefig("Minutes-to-age-distribution")
```

# 6.Goals scored different age categories
#### This visualization compares the total goals scored across different age categories, highlighting how scoring contributions vary with player age. It provides insights into the impact of youth, prime-age, and veteran players on overall team performance. As we can see in the below visualiztion the goals scored by the players of age category 21 to 28 has goals more than 10 on an average. From which we can get to know that the forwards at this age category are more probable to score more than 10 goals in the premier league.

```python
player_age= df[["Age","Gls"]]
player_age
```

```python
fig,ax=plt.subplots(figsize=(7,6))
ax.bar(player_age["Age"],player_age["Gls"],alpha=0.6)
ax.set_xlabel("AGE OF THE PLAYERS",font="Bahnschrift")
ax.set_ylabel("GOALS SCORED",font="Bahnschrift")
ax.legend(["goals scored respective to age"])
ax.axvline(19,color="red",ls="--")
ax.axvline(28.9,color="red",ls="--")
ax.text(29,20,"<- more than 10 goals on \n  average is scored by\n  the age category \n  between 20 and 28",size=15, fontname="Impact",color="blue")

plt.savefig("Goals-by-Age-Category")
```

# 7.Progressive Passes Vs Progressive Carries

This also includes the progressive runs of players with the size of the data points. Players positioned toward the top right excel in both progressing the ball with their feet and distributing it forward effectively. This visualization helps identify versatile playmakers, ball‑carrying midfielders, and deep‑lying distributors who are key to building attacking moves.
